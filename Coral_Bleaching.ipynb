```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#We read our files using pandas
df_fiji=pd.read_csv("Fiji_Bleaching.csv")
df_flor=pd.read_csv("Florida_Keys_Bleaching.csv")
df_island=pd.read_csv("Galapagos_Islands_Bleaching.csv")
df_hawa=pd.read_csv("Hawaii_Bleaching.csv")
df_barrier=pd.read_csv("Northern_Great_Barrier_Reef_Bleaching.csv")
df_palau=pd.read_csv("Palau_Bleaching.csv")

df_flor.head()
```

```python
#Grouped the dataframe by year, then found the mean SST of each year
#Converted the averages into a new CSV file

group_year = df_fiji["Sea_Surface_Temperature"].groupby(df_fiji["Year"]).mean()
group_year.to_csv("Fiji_Mean",header=["Temperature"])

group_year2 = df_flor["Sea_Surface_Temperature"].groupby(df_flor["Year"]).mean()
group_year2.to_csv("Florida_Mean",header=["Temperature"])

group_year3 = df_island["Sea_Surface_Temperature"].groupby(df_island["Year"]).mean()
group_year3.to_csv("Galapagos_Islands_Mean",header=["Temperature"])

group_year4 = df_hawa["Sea_Surface_Temperature"].groupby(df_hawa["Year"]).mean()
group_year4.to_csv("Hawaii_Mean",header=["Temperature"])

group_year5 = df_barrier["Sea_Surface_Temperature"].groupby(df_barrier["Year"]).mean()
group_year5.to_csv("Great_Barrier_Reef_Mean", header=["Temperature"])

group_year6 = df_palau["Sea_Surface_Temperature"].groupby(df_palau["Year"]).mean()
group_year6.to_csv("Palau_Mean",header=["Temperature"])
```

```python
#Read the new CSV files with all the averages
fiji_year=pd.read_csv("Fiji_Mean")

flor_year=pd.read_csv("Florida_Mean")
island_year=pd.read_csv("Galapagos_Islands_Mean")
haw_year=pd.read_csv("Hawaii_Mean")
barrier_year=pd.read_csv("Great_Barrier_Reef_Mean")
palau_year=pd.read_csv("Palau_Mean")

flor_year.head()
```

```python
#Separated the columns
year=flor_year["Year"]

temp_fiji=fiji_year["Temperature"]
temp_flor=flor_year["Temperature"]
temp_island=island_year["Temperature"]
temp_haw=haw_year["Temperature"]
temp_barrier=barrier_year["Temperature"]
temp_palau=palau_year["Temperature"]
```

```python
#Graphed six different lines for each place, with labels and a legend
plt.plot(year, temp_fiji, color="orange")
plt.plot(year, temp_flor, color="blue")
plt.plot(year, temp_island,color="lime")
plt.plot(year, temp_haw,color="darkgray")
plt.plot(year, temp_barrier,color="cyan")
plt.plot(year, temp_palau,color="purple")
plt.legend(labels=["Fiji","Florida Keys","Galapagos Islands","Hawaii","Northern Great Barrier Reef","Palau"], bbox_to_anchor=(1, 1))
plt.title("Sea Surface Temperature Per Year")
plt.grid()
plt.xlabel("Year")
plt.ylabel("Sea Surface Temperature (C)")
plt.savefig("SST_Per_Year.png", bbox_inches="tight")
plt.show()
```

```python
#Graphed six different lines for each place, with labels and a legend
plt.plot(year, temp_fiji, marker='s',color="darkgray",ms=10)
plt.plot(year, temp_flor, marker='s',color="blue",ms=10)
plt.plot(year, temp_island,marker='s',color="lime",ms=10)
plt.plot(year, temp_haw,marker='s',color="orange",ms=10)
plt.plot(year, temp_barrier,marker='s',color="cyan",ms=10)
plt.plot(year, temp_palau,marker='s',color="purple",ms=10)
plt.legend(bbox_to_anchor=(1, 1), labels=["Fiji","Florida Keys","Galapagos Islands","Hawaii","Northern Great Barrier Reef","Palau"])
plt.title("Average Yearly Sea Surface Temperature")
plt.grid()
plt.xlabel("Year")
plt.ylabel("Sea Surface Temperature ($^\circ$C)")
plt.savefig("SST_Per_Year.png", bbox_inches="tight")
plt.show()
```
