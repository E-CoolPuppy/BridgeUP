Change the size of your markers based on a second characteristic

```python
def sizeFunc (mylist, column):
    size=[]
    for item in column: 
        if item <= mylist[0]:
            size.append(10)
        elif item <= mylist[1]:
            size.append(50)
        elif item <= mylist[2]:
            size.append(100)
        elif item <= mylist[3]:
            size.append(150)
        elif item <= mylist[4]:
            size.append(200)
    return size

sizemap = sizeFunc(colorFunc(df["DEATHS"],5), df["DEATHS"])
```

```python
image = plt.imread("earth.jpg")
fig, ax = plt.subplots()
ax.imshow(image, extent = [-180,180,-90,90])

plt.scatter(longitude,latitude, s=sizemap, c=colormap)
plt.xlim(80,180)
plt.ylim(-70,30)
plt.title("Central South Pacific Earthquakes")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.savefig("Central Pacific Earthquakes with Colormap on Earth.png")
plt.show()
```
