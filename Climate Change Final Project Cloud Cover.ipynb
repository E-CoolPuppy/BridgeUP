```python
#Import all the libraries first, so we don't have to worry about it later
import matplotlib.pyplot as plt
import numpy as np
import csv 
```

```python
csvFile = open("CloudFraction1_2002.csv", 'r') 
csvReader = csv.reader(csvFile, delimiter=',')
data_lists = list(csvReader)
```

```python
print(data_lists[0])
```

```python
print(len(data_lists[0]))
```

```python
print(len(data_lists))
```

```python
width=len(data_lists[0])
height=len(data_lists)
```

```python
grid = np.empty([height, width, 3], dtype=np.uint8) 
```

```python
all_values = [] # We will fill in this empty list with all the data values as we go

for r, row in enumerate(data_lists):
    for c, value in enumerate(row):
        if float(value) < 99999:
            all_values.append(float(value)) # This is where we fill in the list with all data values
        if float(value) <= 0: 
            grid[r, c] = [0, 97, 255] #Change [R,G,B]
        elif float(value) <= .1:
            grid[r, c] = [27, 113, 252]
        elif float(value) <= .2:
            grid[r, c] = [66, 138, 255]
        elif float(value) <= .3:
            grid[r, c] = [100, 158, 252]
        elif float(value) <= .4:
            grid[r, c] = [119, 171, 255]
        elif float(value) <= .5:
            grid[r, c] = [140, 184, 255]
        elif float(value) <= .6:
            grid[r, c] = [163, 198, 255]
        elif float(value) <= .7:
            grid[r, c] = [183, 210, 255]
        elif float(value) <= .8:
            grid[r, c] = [206, 224, 255]
        elif float(value) <= .9:
            grid[r, c] = [219, 232, 255]
        elif float(value) <= 1.0:
            grid[r, c] = [255, 255, 255]
        else:
            grid[r, c] = [0, 0, 0]
        
```

```python
print(min(all_values))
print(max(all_values))
#Cannot print min/max of data lists because it is still a lists of list (just prints out the list)
#Must combine it into one long list
```

```python
plt.title("Cloud Fraction in 2002")
plt.imshow(grid)
plt.axis('off')
plt.savefig("Cloud_Fraction_in_2002.png", dpi = 1000)
plt.show()
```

```python
df=pd.read_csv("Annual_NY_CloudCover_csv.csv")
df_clean=df.dropna()
```

```python
df_clean.head()
```

```python
year=df_clean["Year"]
cloud_cover=df_clean["Total Cloud Cover"]
temperature=df_clean["Temperature �"]
radiation=df_clean["Daily Shortwave Radiation"]
```

```python
print(year.head())
print(cloud_cover.head())
print(temperature.head())
print(radiation.head())
```

```python
print(year.min())
print(year.max())
```

```python
plt.plot(year, temperature, color="blue")
plt.legend(loc=4)
plt.twinx()
plt.plot(year, cloud_cover, color="red")
plt.title("Annual NY Cloud Cover")
plt.ylabel("Total Cloud Cover")
plt.legend()
plt.ylabel("Temperature")
plt.xlim(year.min(), year.max())
plt.xticks(range(year.min(), year.max(), 15))
plt.savefig("Annual_NY_Cloud_Cover.png")
plt.show()
#Put twinx after the first plot
```

```python
plt.plot(year, radiation, color="purple")
plt.title("Radiation Each Year")
plt.legend()
plt.ylabel("Radiation")
plt.xlim(year.min(), year.max())
plt.xticks(range(year.min(), year.max(), 15))
plt.savefig("Radiation.png")
plt.show()
```
